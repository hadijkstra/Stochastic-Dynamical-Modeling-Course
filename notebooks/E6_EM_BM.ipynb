# Stochastic Dynamical Models

### Exercise 6: Brownian motion

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(100)

sigma = 0.25; Xzero=1
T=100; N=2**12; dt = float(T)/N
t=np.linspace(0,T,N+1)

M = 1000
Xem=np.zeros((N+1,M+1)); Xem[0,0] = Xzero

for i in range(1,M+1): 
    dW=np.sqrt(dt)*np.random.randn(1,N)
    W=np.cumsum(dW)

    for j in range(1,N+1):
        Winc=np.sum(dW[0][range(j-1,j)])
        Xem[j,i] = Xem[j-1,i] + sigma*Winc
    
    plt.plot(t,Xem[:,i],color='black')
    
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel(r'$t$',fontsize=16); 
plt.ylabel(r'$X$',fontsize=16,rotation=90)
plt.show()
```

```python
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.hist(Xem[100,:], bins = 10, density=True,color='black')
plt.xlabel(r'$X$',fontsize=16); 
plt.ylabel(r'frequency',fontsize=16,rotation=90)
plt.show()
```
