# Stochastic Dynamical Models

Exercise 2: Wiener process 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T=100; N=20000; dt=float(T)/N; 
t=np.linspace(0,T,N+1)
```

```python
dW=np.sqrt(dt)*np.random.normal(0.0,1.0,N)
W=np.cumsum(dW);
W=np.insert(W,0,0)
```

```python
plt.plot(t,W)
plt.xlabel(r'$t$',fontsize=16); plt.ylabel(r'$W_t$',fontsize=16,rotation=0)
plt.title('A Wiener process path')
plt.show()
```
