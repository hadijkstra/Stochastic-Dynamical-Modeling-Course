# Stochastic Dynamical Models

Exercise 1: Gaussian process 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T=100; N=10000;  
t=np.linspace(0,T,N)
```

```python
G = np.random.normal(0.0,1.0,N)
```

```python
plt.plot(t,G)
plt.xlabel(r'$t$',fontsize=16); plt.ylabel(r'$G_t$',fontsize=16,rotation=0)
plt.title('A Gaussian process path')
plt.show()
```

```python
import scipy
from scipy.stats import skew,kurtosis
print(np.mean(G))
print(np.var(G))
print(skew(G, axis=0, bias=True))
print(kurtosis(G, axis=0, fisher=False, bias=True))
```
